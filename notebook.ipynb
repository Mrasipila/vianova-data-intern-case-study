```python
import requests as r
import sqlite3
import pandas as pd
pd.set_option('display.max_rows', 246)
```

# Fetching the data from the API

```python
url = "https://public.opendatasoft.com/api/explore/v2.0/catalog/datasets/geonames-all-cities-with-a-population-1000/exports/csv?delimiter=%3B&list_separator=%2C&quote_all=false&with_bom=false"
res = r.get(url, allow_redirects=True)
open('geonames.csv', 'wb').write(res.content)
```

```python
df = pd.read_csv("geonames.csv", delimiter=";")
df.head()
```

```python
df = df[["geoname_id","cou_name_en","country_code","population"]]
```

# Create the database and store the data inside

```python
db = sqlite3.connect("vianova")
df.to_sql('MAIN', db,index=False, if_exists='replace')
```

# Performing the query : showing the resulting data

```python
latest = pd.read_sql_query("""select country_code, cou_name_en from MAIN where population < 10000000 group by country_code order by cou_name_en """,db)
latest
```
